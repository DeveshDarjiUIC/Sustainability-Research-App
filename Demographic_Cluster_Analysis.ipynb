```python
from google.colab import files

# Upload a file from your local system
uploaded = files.upload()

# Optional: Access the uploaded file as a DataFrame (if it's a CSV)
import pandas as pd

# Get the filename from the uploaded dictionary
filename = list(uploaded.keys())[0]
df = pd.read_csv(filename)

# Display the DataFrame
df.head()
```

```python
df = df.drop(index=27).reset_index(drop=True)
```

```python
from sklearn.cluster import KMeans

population_data = df['Population'].values.reshape(-1, 1)


kmeans = KMeans(n_clusters=3, random_state=42, n_init='auto') # n_init='auto' suppresses a future warning
kmeans.fit(population_data)

# Step 4: Add the cluster labels to the DataFrame
# The cluster assignments are stored in kmeans.labels_
df['Population Cluster'] = kmeans.labels_

# Display the DataFrame with the new cluster column
print("DataFrame with Population Clusters:")
print(df)

# You can also check the cluster centers (mean population for each cluster)
print("\nCluster Centers (Mean Population for each cluster):")
print(kmeans.cluster_centers_)
```

```python
import pandas as pd
from sklearn.preprocessing import StandardScaler
import statsmodels.formula.api as smf
import numpy as np



feature_columns = ['Population', 'Mean Income', 'Population with Bachelors Degree and more', 'Revenue']
X = df[feature_columns]
y = df['Cluster']

# --- 2. Normalize the Independent Variables ---
# Initialize the StandardScaler
scaler = StandardScaler()

# Fit the scaler to your features and transform them
X_scaled = scaler.fit_transform(X)

# Convert the scaled array back to a DataFrame with original column names
X_scaled_df = pd.DataFrame(X_scaled, columns=feature_columns, index=df.index)

# --- 3. Combine Scaled Features with Dependent Variable for statsmodels ---
# statsmodels formula API works well with a single DataFrame
df_scaled = pd.concat([X_scaled_df, y], axis=1)

# For statsmodels, it's often easier if column names don't have spaces or special characters.
# Let's rename them for the formula.
df_scaled.rename(columns={
    'Mean Income': 'Mean_Income',
    'Population with Bachelors Degree and more': 'Pop_Bachelors_Plus'
}, inplace=True)

# Update feature_columns list if you renamed them
formula_feature_columns = ['Population', 'Mean_Income', 'Pop_Bachelors_Plus', 'Revenue']

# --- 4. Run Multinomial Logistic Regression ---
# Create the formula for the model.
# The model will predict 'Cluster' based on the scaled features.
# 'Cluster' is the dependent variable, and the rest are independent variables.
# statsmodels will automatically use the lowest value of 'Cluster' (i.e., 0) as the reference category.
formula = 'Cluster ~ ' + ' + '.join(formula_feature_columns)

# Fit the multinomial logistic regression model
# The `data=df_scaled` argument tells the model where to find the variables.
# `MNLogit` is the class for Multinomial Logit models.
model = smf.mnlogit(formula, data=df_scaled)
results = model.fit()

# --- 5. Display the Summary Table ---
# The summary table contains coefficients, standard errors, z-scores, p-values, and confidence intervals.
print("Multinomial Logistic Regression Results Summary:")
print(results.summary())

# --- 6. Extracting Coefficients and P-values specifically ---
# The summary() method is great, but if you want to programmatically access them:
# The parameters are organized by the outcome category.
# For example, params for predicting Cluster '1' vs Cluster '0' (reference)
# and Cluster '2' vs Cluster '0' (reference).

print("\n--- Coefficients ---")
print(results.params)

print("\n--- P-Values ---")
print(results.pvalues)

```
